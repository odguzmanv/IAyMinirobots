```python
# Lista de entidades
num_entidades = 50
entidades = [f"Entidad_{i+1}" for i in range(num_entidades)]

# Peso político de cada entidad (entre 1 y 100)
pesos_politicos = np.random.randint(1, 101, size=num_entidades)

# Lista de entidades con sus respectivos pesos
entidades_pesos = list(zip(entidades, pesos_politicos))
print("Entidades y sus pesos políticos:", entidades_pesos)

# Número de curules y partidos
num_curules = 50
num_partidos = 5

# Distribución no uniforme de curules entre los partidos
np.random.seed(0)  # Para reproducibilidad
proporciones = np.random.dirichlet(np.ones(num_partidos) * 3)
curules_por_partido = np.round(proporciones * num_curules).astype(int)

# Asegurarnos de que sumen 50 curules
while sum(curules_por_partido) != num_curules:
    curules_por_partido[np.argmin(curules_por_partido)] += 1

print("Curules por partido:", curules_por_partido)

# Asignar las entidades a los partidos según la distribución de curules
partidos = {f"Partido_{i+1}": [] for i in range(num_partidos)}

# Distribuir entidades a los partidos
entidades_indices = list(range(num_entidades))
np.random.shuffle(entidades_indices)

current_index = 0
for i, num_curules in enumerate(curules_por_partido):
    partido = f"Partido_{i+1}"
    for _ in range(num_curules):
        entidad_index = entidades_indices[current_index]
        partidos[partido].append(entidades_pesos[entidad_index])
        current_index += 1

# Mostrar la distribución de entidades por partido
for partido, entidades in partidos.items():
    print(f"{partido}:")
    for entidad, peso in entidades:
        print(f"  {entidad} (Peso: {peso})")

# Calcular el poder total de cada partido
poder_partidos = {partido: sum(peso for entidad, peso in entidades) for partido, entidades in partidos.items()}

# Mostrar el poder total de cada partido
print("\nPoder total por partido:")
for partido, poder in poder_partidos.items():
    print(f"{partido}: {poder} puntos")
```

```python
import random

def fitness(distribucion, curules_por_partido):
    poder_total_partido = np.sum(distribucion * pesos_politicos[:, None], axis=0)
    poder_ideal_partido = (curules_por_partido / num_curules) * np.sum(pesos_politicos)
    error = np.sum(np.abs(poder_total_partido - poder_ideal_partido))
    return -error  # Maximizamos el negativo del error para minimizar el error

def seleccionar_poblacion(poblacion, fitnesses, num_seleccion):
    seleccionados = np.random.choice(len(poblacion), size=num_seleccion, p=fitnesses/np.sum(fitnesses))
    return [poblacion[i] for i in seleccionados]

def cruzamiento(padre1, padre2):
    punto_corte = np.random.randint(1, num_entidades)
    hijo1 = np.vstack((padre1[:punto_corte], padre2[punto_corte:]))
    hijo2 = np.vstack((padre2[:punto_corte], padre1[punto_corte:]))
    return hijo1, hijo2

def mutacion(distribucion, tasa_mutacion=0.01):
    for i in range(len(distribucion)):
        if random.random() < tasa_mutacion:
            partido = np.random.randint(num_partidos)
            distribucion[i] = np.eye(num_partidos)[partido]
    return distribucion
```

```python
poblacion_size = 100
poblacion = []
for _ in range(poblacion_size):
    distribucion = np.eye(num_partidos)[np.random.choice(num_partidos, num_entidades)]
    poblacion.append(distribucion)
```

```python
num_generaciones = 1000
tasa_cruzamiento = 0.7
tasa_mutacion = 0.01

mejor_fitness = float('-inf')
mejor_distribucion = None

for generacion in range(num_generaciones):
    fitnesses = np.array([fitness(individuo, curules_por_partido) for individuo in poblacion])

    if max(fitnesses) > mejor_fitness:
        mejor_fitness = max(fitnesses)
        mejor_distribucion = poblacion[np.argmax(fitnesses)]

    nueva_poblacion = []
    while len(nueva_poblacion) < poblacion_size:
        padres = seleccionar_poblacion(poblacion, fitnesses, 2)
        if random.random() < tasa_cruzamiento:
            hijo1, hijo2 = cruzamiento(padres[0], padres[1])
        else:
            hijo1, hijo2 = padres[0], padres[1]

        hijo1 = mutacion(hijo1, tasa_mutacion)
        hijo2 = mutacion(hijo2, tasa_mutacion)

        nueva_poblacion.append(hijo1)
        nueva_poblacion.append(hijo2)

    poblacion = nueva_poblacion[:poblacion_size]

print("Mejor distribución encontrada:")
print(mejor_distribucion)
```

```python
import numpy as np

# Datos del problema
capacidad_plantas = np.array([3, 6, 5, 4])  # Capacidad de generación de cada planta
demanda_ciudades = np.array([4, 3, 5, 3])  # Demanda de energía de cada ciudad
costos_transporte = np.array([[0, 4, 3, 6],
                               [4, 0, 4, 5],
                               [3, 4, 0, 4],
                               [6, 5, 4, 0]])  # Costos de transporte por GW entre plantas y ciudades
costos_generacion = np.array([680, 720, 660, 750])  # Costos por KW-H de cada generador

# Función para calcular el costo total de transporte
def calcular_costo_transporte(solucion):
    costo = 0
    for planta in range(len(solucion)):
        ciudad = solucion[planta]
        costo += costos_transporte[planta][ciudad]
    return costo

# Función para calcular el costo total de generación
def calcular_costo_generacion(solucion):
    costo = 0
    for planta, generador in enumerate(solucion):
        costo += costos_generacion[planta] * capacidad_plantas[planta] * (1 if generador == planta else 0)
    return costo

# Función de aptitud: minimizar costo total de transporte y generación
def calcular_aptitud(solucion):
    costo_transporte = calcular_costo_transporte(solucion)
    costo_generacion = calcular_costo_generacion(solucion)
    return costo_transporte + costo_generacion
```

```python
from random import randint, random

# Función para generar una solución aleatoria
def generar_solucion():
    return [randint(0, 3) for _ in range(4)]  # 4 ciudades, 4 plantas

# Crear población inicial
poblacion_size = 50
poblacion = [generar_solucion() for _ in range(poblacion_size)]

# Parámetros del AG
num_generaciones = 100
tasa_mutacion = 0.1
```

```python
# Función de selección por torneo
def seleccion_torneo(poblacion, aptitudes):
    seleccionados = []
    for _ in range(poblacion_size):
        idx1, idx2 = randint(0, poblacion_size-1), randint(0, poblacion_size-1)
        seleccionados.append(poblacion[idx1] if aptitudes[idx1] < aptitudes[idx2] else poblacion[idx2])
    return seleccionados

# Función de cruce (crossover)
def crossover(padre, madre):
    punto_corte = randint(1, len(padre) - 1)
    hijo = padre[:punto_corte] + madre[punto_corte:]
    return hijo

# Algoritmo genético
mejor_solucion = None
mejor_aptitud = float('inf')

for generacion in range(num_generaciones):
    # Calcular aptitud de la población
    aptitudes = [calcular_aptitud(individuo) for individuo in poblacion]

    # Selección
    seleccionados = seleccion_torneo(poblacion, aptitudes)

    # Cruzamiento
    nueva_poblacion = []
    for i in range(0, poblacion_size, 2):
        padre, madre = seleccionados[i], seleccionados[i + 1]
        hijo1 = crossover(padre, madre)
        hijo2 = crossover(madre, padre)
        nueva_poblacion.append(hijo1)
        nueva_poblacion.append(hijo2)

    # Mutación
    for individuo in nueva_poblacion:
        if random() < tasa_mutacion:
            idx = randint(0, len(individuo) - 1)
            individuo[idx] = randint(0, 3)

    poblacion = nueva_poblacion

    # Actualizar mejor solución
    mejor_idx = aptitudes.index(min(aptitudes))
    if aptitudes[mejor_idx] < mejor_aptitud and sum(poblacion[mejor_idx]) > 0:
        mejor_solucion = poblacion[mejor_idx]
        mejor_aptitud = aptitudes[mejor_idx]

print("Mejor solución encontrada:", mejor_solucion)
print("Costo total:", calcular_aptitud(mejor_solucion))
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Parámetros
n_cities = 20
n_population = 100
mutation_rate = 0.3
n_generations = 1000  # Puedes ajustar este valor según sea necesario
selection_type = "proportional"  # Cambia a "tournament" si deseas usar selección por torneo
```

```python
# Generar la lista de coordenadas representando cada ciudad
coordinates_list = [[x,y] for x,y in zip(np.random.randint(0,100,n_cities),np.random.randint(0,100,n_cities))]
names_list = np.array(['Berlin', 'London', 'Moscow', 'Barcelona', 'Rome', 'Paris', 'Vienna', 'Munich', 'Istanbul', 'Kyiv', 'Bucharest', 'Minsk', 'Warsaw', 'Budapest', 'Milan', 'Prague', 'Sofia', 'Birmingham', 'Brussels', 'Amsterdam'])
cities_dict = { x:y for x,y in zip(names_list,coordinates_list)}
```

```python
# Función para calcular la distancia entre dos puntos
def compute_city_distance_coordinates(a,b):
    return ((a[0]-b[0])**2+(a[1]-b[1])**2)**0.5

def compute_city_distance_names(city_a, city_b, cities_dict):
    return compute_city_distance_coordinates(cities_dict[city_a], cities_dict[city_b])
```

```python
# Generar la primera población
def genesis(city_list, n_population):
    population_set = []
    for i in range(n_population):
        sol_i = city_list[np.random.choice(list(range(n_cities)), n_cities, replace=False)]
        population_set.append(sol_i)
    return np.array(population_set)
```

```python
# Función para evaluar la aptitud de una solución
def fitness_eval(city_list, cities_dict):
    total = 0
    for i in range(n_cities-1):
        a = city_list[i]
        b = city_list[i+1]
        total += compute_city_distance_names(a, b, cities_dict)
    total += compute_city_distance_names(city_list[-1], city_list[0], cities_dict)
    return total
```

```python
# Evaluar la aptitud de toda la población
def get_all_fitnes(population_set, cities_dict):
    fitnes_list = []
    for solution in population_set:
        fitness = fitness_eval(solution, cities_dict)
        fitnes_list.append(fitness)
    return np.array(fitnes_list)
```

```python
# Selección de progenitores
def progenitor_selection(population_set, fitnes_list):
    if selection_type == "proportional":
        total_fit = fitnes_list.sum()
        prob_list = fitnes_list / total_fit
        progenitor_list_a = np.random.choice(
            list(range(len(population_set))),
            len(population_set),
            p=prob_list,
            replace=True,
        )
        progenitor_list_b = np.random.choice(
            list(range(len(population_set))),
            len(population_set),
            p=prob_list,
            replace=True,
        )
    elif selection_type == "tournament":
        progenitor_list_a = np.random.randint(
            0, len(population_set), len(population_set)
        )
        progenitor_list_b = np.random.randint(
            0, len(population_set), len(population_set)
        )

    progenitor_list_a = population_set[progenitor_list_a]
    progenitor_list_b = population_set[progenitor_list_b]

    return np.array([progenitor_list_a, progenitor_list_b])
```

```python
# Cruzar progenitores para crear descendencia
def mate_progenitors(prog_a, prog_b):
    offspring = prog_a[0:5]
    for city in prog_b:
        if city not in offspring:
            offspring = np.concatenate((offspring, [city]))
    return offspring

def mate_population(progenitor_list):
    new_population_set = []
    for i in range(progenitor_list.shape[1]):
        prog_a, prog_b = progenitor_list[0][i], progenitor_list[1][i]
        offspring = mate_progenitors(prog_a, prog_b)
        new_population_set.append(offspring)
    return new_population_set
```

```python
# Mutar descendencia
def mutate_offspring(offspring):
    for q in range(int(n_cities * mutation_rate)):
        a = np.random.randint(0, n_cities)
        b = np.random.randint(0, n_cities)
        offspring[a], offspring[b] = offspring[b], offspring[a]
    return offspring

def mutate_population(new_population_set):
    mutated_pop = []
    for offspring in new_population_set:
        mutated_pop.append(mutate_offspring(offspring))
    return mutated_pop
```

```python
# Función para imprimir las gráficas
def plot_results(best_solution_history, fitness_history):
    plt.figure(figsize=(12, 6))

    # Gráfica de evolución de la función de fitness
    plt.subplot(1, 2, 1)
    plt.plot(range(len(fitness_history)), fitness_history, label="Fitness")
    plt.xlabel("Generations")
    plt.ylabel("Fitness")
    plt.title("Fitness Evolution")
    plt.legend()

    # Gráfica del mejor recorrido encontrado
    plt.subplot(1, 2, 2)
    plt.plot(
        [cities_dict[city][0] for city in best_solution_history[2]],
        [cities_dict[city][1] for city in best_solution_history[2]],
        label="Best Solution",
        marker="o",
    )
    plt.xlabel("X Coordinates")
    plt.ylabel("Y Coordinates")
    plt.title("Best Solution")
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
# Generación de la población inicial
population_set = genesis(names_list, n_population)

# Evaluación de la aptitud (fitness) de la población inicial
fitnes_list = get_all_fitnes(population_set, cities_dict)

# Inicialización de las variables para guardar el historial de la mejor solución y la aptitud
best_solution_history = [[], [], []]
fitness_history = []
```

```python
# Bucle principal del algoritmo genético
for i in range(n_generations):
    if i % 100 == 0:
        print(
            i,
            fitnes_list.min(),
            fitnes_list.mean(),
            datetime.now().strftime("%d/%m/%y %H:%M"),
        )

    # Guardar la aptitud mínima de la generación actual
    fitness_history.append(fitnes_list.min())

    # Guardar la mejor solución encontrada
    if len(best_solution_history[1]) == 0 or fitnes_list.min() < best_solution_history[1][-1]:
        best_solution_history[0].append(i)
        best_solution_history[1].append(fitnes_list.min())
        best_solution_history[2] = population_set[fitnes_list.argmin()]

    # Selección de progenitores y generación de nueva población
    progenitor_list = progenitor_selection(population_set, fitnes_list)
    new_population_set = mate_population(progenitor_list)

    # Mutación de la población
    mutated_pop = mutate_population(new_population_set)

    # Evaluación de la aptitud de la nueva población
    fitnes_list = get_all_fitnes(mutated_pop, cities_dict)
```

```python
# Imprimir las gráficas
plot_results(best_solution_history, fitness_history)
```

```python
# Importar las bibliotecas necesarias
import numpy as np
import random
import string
from gtts import gTTS
from IPython.display import Audio, display
```

```python
# Función para generar una palabra aleatoria de una longitud dada
def generate_random_word(length=5): #utilizando letras minúsculas.
    return ''.join(random.choices(string.ascii_lowercase, k=length))

# Función para calcular la aptitud (similaridad con la palabra objetivo)
def fitness(word, target_word): #contando el número de caracteres coincidentes.
    return sum(c1 == c2 for c1, c2 in zip(word, target_word))

# Función para crear la población inicial de palabras aleatorias
def create_population(size, word_length=5): #Crea una lista de palabras aleatorias para iniciar la población.
    return [generate_random_word(word_length) for _ in range(size)]

# Función para seleccionar progenitores usando el método de torneo
def tournament_selection(population, fitnesses, k=3): #donde se elige el mejor de un grupo aleatorio de candidatos.
    selected = []
    for _ in range(len(population)):
        aspirants = random.sample(list(zip(population, fitnesses)), k)
        selected.append(max(aspirants, key=lambda x: x[1])[0])
    return selected

# Función para cruzar dos progenitores y generar descendencia
def crossover(parent1, parent2): #Combina dos palabras progenitoras en un punto de corte aleatorio para generar descendencia.
    point = random.randint(1, len(parent1) - 1)
    return parent1[:point] + parent2[point:]

# Función para mutar una palabra con una tasa de mutación dada cambiando letras aleatorias.
def mutate(word, mutation_rate=0.1):
    word = list(word)
    for i in range(len(word)):
        if random.random() < mutation_rate:
            word[i] = random.choice(string.ascii_lowercase)
    return ''.join(word)

# Función para reproducir una palabra usando gTTS y reproducir el audio
def speak_word(word):
    tts = gTTS(text=word, lang='en')
    tts.save("word.mp3")
    display(Audio("word.mp3", autoplay=True))

# Función principal del algoritmo genético
# Ejecuta el algoritmo genético, iterando a través de generaciones para encontrar la palabra objetivo.
# Muestra y reproduce la mejor palabra cada 100 generaciones y al final del proceso.
def genetic_algorithm(target_word, population_size=50, generations=1000, mutation_rate=0.1):
    word_length = len(target_word)
    population = create_population(population_size, word_length)

    best_solution_history = [[], []]
    fitness_history = []

    for generation in range(generations):
        fitnesses = [fitness(word, target_word) for word in population]

        # Si se encuentra la palabra objetivo, se termina la ejecución
        if max(fitnesses) == word_length:
            break

        # Selección de progenitores
        selected = tournament_selection(population, fitnesses)
        next_population = []

        # Generación de la siguiente población
        for i in range(0, population_size, 2):
            parent1 = selected[i]
            parent2 = selected[i + 1]
            child1 = crossover(parent1, parent2)
            child2 = crossover(parent2, parent1)
            next_population.extend([mutate(child1, mutation_rate), mutate(child2, mutation_rate)])

        population = next_population

        # Cada 100 generaciones, mostrar la mejor palabra y reproducirla
        if generation % 100 == 0:
            best_word = population[np.argmax(fitnesses)]
            print(f"Generation {generation}: Best word = {best_word}, Fitness = {max(fitnesses)}")
            speak_word(best_word)

    # Mostrar y reproducir la mejor palabra al final de todas las generaciones
    best_word = population[np.argmax(fitnesses)]
    print(f"Final Generation: Best word = {best_word}, Fitness = {max(fitnesses)}")
    speak_word(best_word)
    return best_word
```

```python
# Definir la Palabra objetivo
target_word = "minirobots"
```

```python
# Ejecutar el algoritmo genético
best_word = genetic_algorithm(target_word)

# Imprimir la mejor palabra encontrada
print(f"Best word found: {best_word}")
```

El código utiliza un algoritmo genético para encontrar una palabra específica a partir de un grupo de palabras generadas aleatoriamente. Este algoritmo sigue varios pasos: primero, genera una población inicial de palabras aleatorias. Luego, evalúa qué tan cerca está cada palabra de la palabra objetivo usando una función de aptitud que mide la similitud entre las palabras.

La aptitud es una medida de qué tan bien una palabra generada se parece a la palabra objetivo. Se calcula contando cuántas letras coinciden en la posición correcta entre la palabra generada y la palabra objetivo. Cuantas más letras coincidan, mayor es la aptitud de la palabra.

Después de evaluar la aptitud, selecciona las mejores palabras (las que más se parecen a la palabra objetivo) para que sean los "padres" de la próxima generación. Estos padres se mezclan mediante un proceso llamado cruce, creando nuevas palabras que combinan partes de ambos padres.
```python
import numpy as np
import pygad
import matplotlib.pyplot as plt
```

```python
# Generar las coordenadas de las ciudades
n_cities = 20
coordinates_list = [[x, y] for x, y in zip(np.random.randint(0, 100, n_cities), np.random.randint(0, 100, n_cities))]
```

```python
# Calcular la distancia entre dos puntos
def compute_distance(a, b):
    return np.sqrt(np.sum((np.array(a) - np.array(b))**2))
```

```python
# Crear una función de aptitud para evaluar las soluciones
def fitness_func(ga_instance, solution, solution_idx):
    total_distance = 0
    for i in range(n_cities - 1):
        total_distance += compute_distance(coordinates_list[solution[i]], coordinates_list[solution[i + 1]])
    total_distance += compute_distance(coordinates_list[solution[-1]], coordinates_list[solution[0]])  # Regresar al punto de inicio
    return -total_distance  # Queremos minimizar la distancia total
```

```python
# Parámetros del algoritmo genético
num_generations = 1000
num_parents_mating = 10
sol_per_pop = 50
num_genes = n_cities
```

```python
# Crear la instancia de Pygad
ga_instance = pygad.GA(
    num_generations=num_generations,
    num_parents_mating=num_parents_mating,
    fitness_func=fitness_func,
    sol_per_pop=sol_per_pop,
    num_genes=num_genes,
    gene_type=int,
    init_range_low=0,
    init_range_high=n_cities,
    mutation_type="random",
    mutation_percent_genes=10,
    gene_space=list(range(n_cities)),
    crossover_type="single_point",
    parent_selection_type="rank"
)
```

```python
# Ejecutar el algoritmo genético
ga_instance.run()
```

```python
# Obtener la mejor solución
solution, solution_fitness, solution_idx = ga_instance.best_solution()
```

```python
# Imprimir la mejor solución
print(f"Mejor solución: {solution}")
print(f"Distancia total: {-solution_fitness}")
```

```python
# Visualizar el mejor recorrido
solution_coordinates = [coordinates_list[solution[i]] for i in range(n_cities)]
solution_coordinates.append(solution_coordinates[0])  # Regresar al punto de inicio
solution_coordinates = np.array(solution_coordinates)

plt.figure(figsize=(10, 6))
plt.plot(solution_coordinates[:, 0], solution_coordinates[:, 1], marker='o')
plt.title(f"Mejor recorrido (distancia total: {-solution_fitness:.2f})")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```

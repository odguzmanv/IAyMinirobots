1.1 La inteligencia artificial son algoritmos que son ejecutados por una maquina con la intención de realizar tareas para las cuales se les entrena, para esto se le debe alimentar con una gran cantidad de datos. Hay aprendizaje supervisado y no supervisado en el cual su principal diferencia es el conocimiento del programador en cuanto a los valores que tiene el algoritmo para reaccionar a ciertos estimulos, siendo el no supervisado el que varia automaticamente estos pesos sin el conocimiento del programador en cada época de aprendizaje.

  

  > Según ChatGpt al prompt "¿Qué es la inteligencia artificial?": La inteligencia artificial (IA) se refiere a la capacidad de las máquinas o sistemas informáticos para realizar tareas que normalmente requieren inteligencia humana. Esto incluye el aprendizaje, la percepción, el razonamiento, la resolución de problemas y la toma de decisiones. Los sistemas de inteligencia artificial pueden ser diseñados para imitar funciones cognitivas humanas, como el reconocimiento de voz, la visión por computadora, el procesamiento del lenguaje natural y la toma de decisiones basada en datos.

  > Hay diferentes enfoques dentro de la inteligencia artificial, que incluyen la inteligencia artificial débil, que se centra en tareas específicas y no pretende imitar la inteligencia humana en su totalidad, y la inteligencia artificial fuerte, que busca replicar la inteligencia humana en su totalidad. La inteligencia artificial ha encontrado aplicaciones en una amplia gama de campos, incluyendo la medicina, la industria automotriz, las finanzas, la agricultura, la atención al cliente, entre otros.

1.3 Colombia ha presentado una Hoja de Ruta de la Inteligencia Artificial, liderada por la ministra Yesenia Olaya, para guiar el desarrollo ético y sostenible de la IA en el país. Esta estrategia se centra en la transparencia, educación, innovación, colaboración entre industrias y academia, manejo seguro de datos, y ciberseguridad. Incluye iniciativas como un Comité de Expertos en IA y la participación en proyectos internacionales, además de la colaboración con el Ministerio de Tecnologías de la Información para desarrollar un documento CONPES en IA, alineado con el Plan Nacional de Desarrollo 2022-2026.

Perú, mediante la Ley N° 31814 promulgada el 5 de julio de 2023, promueve el uso de la inteligencia artificial para el desarrollo económico y social, garantizando su uso ético, sostenible y transparente. La autoridad encargada de supervisar y promover la IA es la Presidencia del Consejo de Ministros a través de la Secretaría de Gobierno y Transformación Digital. Se estableció la creación de equipos técnicos multidisciplinarios en entidades públicas, el uso del Centro Nacional de Innovación Digital e Inteligencia Artificial y la clasificación de los sistemas de IA según el riesgo que representan para la equidad, transparencia y derechos fundamentales. Además, se fomenta la publicación de algoritmos y datos en plataformas nacionales para su uso abierto y público.

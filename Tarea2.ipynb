### Modelo "Fire" (Fuego) en NetLogo

#### Descripcin del Modelo
El modelo "Fire" simula cmo un incendio se propaga a travs de un bosque. El bosque se representa como una cuadrcula de celdas, donde cada celda puede estar en uno de tres estados:
- rbol (verde)
- rbol en llamas (rojo)
- rbol quemado (negro)

#### Reglas del Modelo
- Un rbol que no est en llamas puede comenzar a arder si al menos uno de sus vecinos est en llamas.
- Un rbol que est en llamas se convertir en un rbol quemado en el siguiente paso de tiempo.
- Un rbol quemado no puede volver a arder.

#### Parmetros Ajustables
- **Densidad de rboles:** La probabilidad de que una celda contenga un rbol al inicio de la simulacin.
- **Probabilidad de propagacin del fuego:** La probabilidad de que un rbol comience a arder si uno de sus vecinos est en llamas.

#### Observaciones y Anlisis

- **Densidad del Bosque:**

  Antes del 60% de Densidad:
La quema porcentual es inferior al 40%, muchas veces no llega al 15%, pero nunca baja del 10%.

  Zona Crtica (57% - 59%):
La velocidad de propagacin es exponencial al principio, pero se estabiliza cuando hay menos caminos para seguir quemando.
Cuando la quema supera el 30%, la propagacin se vuelve ms lenta hasta detenerse.

  Densidad Superior al 60%:
La propagacin es rpida y extensa, con una mayor probabilidad de que el incendio alcance el borde derecho del bosque.

  Comportamiento del Incendio:
El fuego se comporta de manera similar a un resolvedor de laberintos, buscando los caminos disponibles.
Comparado con el hongo Physarum polycephalum, que optimiza el transporte encontrando el camino ms corto entre dos puntos, el fuego tambin muestra un comportamiento optimizador al quemar.

- **Patrones Similares:**

  1. Resolucin de Laberintos: Al igual que el fuego, un hongo como Physarum polycephalum o algoritmos computacionales pueden resolver laberintos optimizando el camino ms corto.
  2. Propagacin de Epidemias: Las epidemias pueden mostrar patrones similares de propagacin en una poblacin densa, donde la tasa de infeccin es alta en densidades crticas.
  3. Difusin de Contaminantes: La difusin de contaminantes en un medio fluido puede seguir patrones parecidos, donde la concentracin y la tasa de dispersin afectan la propagacin.

Se us el automata 30: Segn ChatGPT la regla 30 es una de las reglas ms conocidas en el campo de los autmatas celulares unidimensionales, introducida por el matemtico Stephen Wolfram. Es famosa por generar patrones complejos y caticos a partir de reglas muy simples. En un autmata celular unidimensional, una lnea de celdas puede estar en uno de dos estados: 0 (blanco) o 1 (negro). La evolucin del autmata se realiza en pasos discretos de tiempo, comenzando con una configuracin inicial que suele ser una lnea con una sola celda en el estado 1 (negro) y todas las dems en el estado 0 (blanco).

Cada celda en el autmata celular 1D tiene una vecindad que incluye a s misma y a sus dos vecinos inmediatos, formando un bloque de tres celdas. La regla 30 determina el estado de una celda en el siguiente paso de tiempo en funcin del estado de la celda y sus dos vecinos en el paso de tiempo actual. Existen ocho posibles combinaciones de estados para la vecindad de tres celdas, y cada combinacin se asigna a un nuevo estado (0 o 1) segn la regla 30.

Las combinaciones y sus resultados en la regla 30 son las siguientes: si la vecindad es 111, 110, o 101, el nuevo estado ser 0; si la vecindad es 100, 011, 010, o 001, el nuevo estado ser 1; y si la vecindad es 000, el nuevo estado ser 0. El nombre "regla 30" proviene del hecho de que las salidas "00011110" se interpretan como el nmero binario 30 en decimal.

Supongamos que tenemos una lnea de siete celdas con una sola celda en el estado 1 en el centro. La configuracin inicial es "0001000". Aplicando la regla 30 a cada celda (excepto las de los extremos, que tienen vecinos faltantes), obtenemos la siguiente evolucin paso a paso: "0001000", "0011100", "0110010", "1101111", "0011000".

La regla 30 es conocida por su comportamiento catico y su capacidad de generar patrones complejos a partir de reglas simples. A pesar de la simplicidad de las reglas, la evolucin del sistema puede mostrar una gran riqueza de estructuras y comportamientos, lo que la hace interesante tanto para la teora de la computacin como para la investigacin en sistemas complejos y teora del caos.

La regla 30, al igual que otros autmatas celulares, sirve como un ejemplo de cmo la complejidad puede surgir de reglas simples. Esto tiene implicaciones en muchos campos, incluyendo la fsica, la biologa, la teora de la computacin y ms, mostrando cmo patrones complejos y comportamientos emergentes pueden surgir de interacciones simples y locales.

Se programo el 30 que devuelve un archivo CSV que esta en la carpeta de las tareas y un segundo cdigo que lo convierte en archivo STL que es el que sera interpretado por una impresora, este archivo tambin est en la carpeta de las tareas.

El cdigo y la imagen de como quedara el automata impreso sern mostrados a continuacin. Se uso https://www.viewstl.com/ para la visualizacin del automata en 3D.

```python
import numpy as np

# Definir la regla 30
def regla_30(izquierda, centro, derecha):
    return izquierda ^ (centro | derecha)

# Inicializar la primera fila
ancho = 100
pasos = 50
estado_inicial = np.zeros(ancho, dtype=int)
estado_inicial[ancho // 2] = 1

# Matriz para almacenar los estados
estados = np.zeros((pasos, ancho), dtype=int)
estados[0] = estado_inicial

# Simular el autmata
for i in range(1, pasos):
    for j in range(1, ancho - 1):
        estados[i, j] = regla_30(estados[i - 1, j - 1], estados[i - 1, j], estados[i - 1, j + 1])

# Guardar los estados en un archivo (opcional)
np.savetxt("estados_automata.csv", estados, fmt="%d", delimiter=",")
```

```python
import os
import bpy
import csv

# Definir el tamao de cada cubo
tamano_cubo = 1.0

# Cargar los estados desde el archivo CSV
with open("estados_automata.csv") as csvfile:
    reader = csv.reader(csvfile)
    estados = np.array([[int(celda) for celda in fila] for fila in reader])

# Funcin para crear un cubo en Blender
def crear_cubo(x, y, z):
    bpy.ops.mesh.primitive_cube_add(size=tamano_cubo, location=(x * tamano_cubo, y * tamano_cubo, z * tamano_cubo))

# Crear cubos para cada celda activa
for t, fila in enumerate(estados):
    for x, celda in enumerate(fila):
        if celda == 1:
            crear_cubo(x, t, 0)  # Usar el tiempo como la segunda dimensin espacial

# Verificar si el directorio de salida existe, si no, crearlo
output_dir = "output"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Guardar el modelo como un archivo STL
output_filepath = os.path.join(output_dir, "automata.stl")
bpy.ops.export_mesh.stl(filepath=output_filepath)
```

Comenzamos creando la simulacin en Python y luego un cdigo para contar las casillas en cada estado.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from ipywidgets import interact, IntSlider
import io
from PIL import Image

# Parmetros (puedes ajustarlos segn sea necesario)
N = 100  # Tamao del retculo (lattice)
ciclos = 100  # Nmero de ciclos de simulacin
ps = 0.4  # Probabilidad de infeccin de una persona sana por vecinos sintomticos
pa = 0.3  # Probabilidad de infeccin de una persona sana por vecinos asintomticos
rs = 0.2  # Probabilidad de recuperacin del estado sintomtico en cada ciclo
ra = 0.1  # Probabilidad de recuperacin del estado asintomtico en cada ciclo
md = 0.1  # Probabilidad de muerte en estado sintomtico en cada ciclo
NP, PS, CS, CNS, CT, PAC = range(6)  # ndices de estado para mayor claridad

def init(N, ns, cs, ca):
    """Inicializa el retculo con nmeros especficos de personas."""
    g = np.zeros((N, N), int)
    for _ in range(ns):
        x, y = random.randint(0, N - 1), random.randint(0, N - 1)
        while g[x, y]:
            x, y = random.randint(0, N - 1), random.randint(0, N - 1)
        g[x, y] = PS  # Persona sana
    for _ in range(cs):
        x, y = random.randint(0, N - 1), random.randint(0, N - 1)
        while g[x, y]:
            x, y = random.randint(0, N - 1), random.randint(0, N - 1)
        g[x, y] = CS  # Persona con COVID-19 sintomtica
    for _ in range(ca):
        x, y = random.randint(0, N - 1), random.randint(0, N - 1)
        while g[x, y]:
            x, y = random.randint(0, N - 1), random.randint(0, N - 1)
        g[x, y] = CNS  # Persona con COVID-19 asintomtica
    return g

def step(g, muertos):
    """Simula un paso temporal, actualizando el estado de cada celda."""
    ng = g.copy()
    for x in range(N):
        for y in range(N):
            s = g[x, y]
            if s == PS:  # Persona sana
                n_infected = sum(g[(x + i) % N, (y + j) % N] in [CS, CNS]
                                 for i in [-1, 0, 1] for j in [-1, 0, 1] if (i, j) != (0, 0))
                if n_infected > 0:
                    if random.random() < ps:  # Mayor probabilidad de infeccin por sintomtico
                        ng[x, y] = CS
                    elif random.random() < pa:  # Menor probabilidad de infeccin por asintomtico
                        ng[x, y] = CNS
            elif s == CS:  # Persona con COVID-19 sintomtica
                if random.random() < md:  # Muerte
                    ng[x, y] = NP
                    muertos += 1
                elif random.random() < rs:  # Recuperacin
                    ng[x, y] = PAC
                else:  # Contina sintomtico
                    ng[x, y] = CS
            elif s == CNS:  # Persona con COVID-19 asintomtico
                if random.random() < ra:  # Recuperacin
                    ng[x, y] = PAC
                else:  # Contina asintomtico
                    ng[x, y] = CNS
            elif s == CT:  # En cuarentena o en tratamiento
                if random.random() < rs:
                    ng[x, y] = PAC  # Recupera
                else:
                    ng[x, y] = NP  # Muere
                    muertos += 1
            elif s == PAC:  # Persona curada con anticuerpos
                ng[x, y] = PAC  # Se asume que la inmunidad es persistente
    return ng, muertos

def calcular_porcentajes(g, muertos):
    """Calcula el porcentaje de cada estado en el retculo y el porcentaje de muertos."""
    unique, counts = np.unique(g, return_counts=True)
    total = g.size
    porcentajes = dict(zip(unique, counts / total * 100))
    porcentajes[6] = (muertos / total) * 100  # Aadir el porcentaje de muertos con ndice 6
    return porcentajes

def almacenar_porcentajes(porcentajes):
    """Almacena los porcentajes de cada estado en una cadena formateada."""
    nombres_estados = ['No hay nada', 'Persona sana', 'COVID-19 sintomtica', 'COVID-19 asintomtica', 'Cuidados mdicos', 'Curada con anticuerpos', 'Muertos']
    texto_porcentajes = "\n".join([f'{nombres_estados[estado]}: {porcentaje:.2f}%' for estado, porcentaje in porcentajes.items()])
    return texto_porcentajes

def sim():
    """Ejecuta la simulacin completa."""
    g = init(N, 5000, 5, 5)
    imgs = []  # Lista para almacenar las imgenes de cada ciclo
    porcentajes_info = []  # Lista para almacenar los porcentajes de cada ciclo
    muertos = 0  # Contador de muertos

    for _ in range(ciclos):
        g, muertos = step(g, muertos)

        # Calcular los porcentajes de cada estado y de muertos
        porcentajes = calcular_porcentajes(g, muertos)
        # Almacenar la informacin de los porcentajes en el buffer
        porcentajes_info.append(almacenar_porcentajes(porcentajes))

        # Almacenar imgenes sin mostrarlas individualmente (mejora el rendimiento)
        buf = io.BytesIO()
        plt.imshow(g, cmap='viridis', vmin=0, vmax=5)
        plt.savefig(buf, format='png')
        buf.seek(0)
        imgs.append(Image.open(buf))
        plt.close()

    # Crear el control deslizante interactivo
    def show(i):
        # Mostrar la imagen correspondiente al ciclo i
        plt.imshow(imgs[i], cmap='viridis', vmin=0, vmax=5)
        # Mostrar la barra de color con etiquetas personalizadas equidistantes
        cbar = plt.colorbar(ticks=[0, 1, 2, 3, 4, 5])
        cbar.set_ticklabels(['No hay nada', 'Persona sana', 'COVID-19 sintomtica', 'COVID-19 asintomtica', 'Cuidados mdicos', 'Curada con anticuerpos'])
        # Mostrar la informacin de los porcentajes
        plt.figtext(0.15, 0.85, porcentajes_info[i], fontsize=10, bbox={"facecolor": "white", "alpha": 0.5, "pad": 5})
        plt.show()

    interact(show, i=IntSlider(min=0, max=len(imgs) - 1))

sim()
```

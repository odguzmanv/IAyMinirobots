### Modelo "Fire" (Fuego) en NetLogo

#### Descripción del Modelo
El modelo "Fire" simula cómo un incendio se propaga a través de un bosque. El bosque se representa como una cuadrícula de celdas, donde cada celda puede estar en uno de tres estados:
- Árbol (verde)
- Árbol en llamas (rojo)
- Árbol quemado (negro)

#### Reglas del Modelo
- Un árbol que no está en llamas puede comenzar a arder si al menos uno de sus vecinos está en llamas.
- Un árbol que está en llamas se convertirá en un árbol quemado en el siguiente paso de tiempo.
- Un árbol quemado no puede volver a arder.

#### Parámetros Ajustables
- **Densidad de árboles:** La probabilidad de que una celda contenga un árbol al inicio de la simulación.
- **Probabilidad de propagación del fuego:** La probabilidad de que un árbol comience a arder si uno de sus vecinos está en llamas.

#### Observaciones y Análisis

- **Densidad del Bosque:**

  Antes del 60% de Densidad:
La quema porcentual es inferior al 40%, muchas veces no llega al 15%, pero nunca baja del 10%.

  Zona Crítica (57% - 59%):
La velocidad de propagación es exponencial al principio, pero se estabiliza cuando hay menos caminos para seguir quemando.
Cuando la quema supera el 30%, la propagación se vuelve más lenta hasta detenerse.

  Densidad Superior al 60%:
La propagación es rápida y extensa, con una mayor probabilidad de que el incendio alcance el borde derecho del bosque.

  Comportamiento del Incendio:
El fuego se comporta de manera similar a un resolvedor de laberintos, buscando los caminos disponibles.
Comparado con el hongo Physarum polycephalum, que optimiza el transporte encontrando el camino más corto entre dos puntos, el fuego también muestra un comportamiento optimizador al quemar.

- **Patrones Similares:**

  1. Resolución de Laberintos: Al igual que el fuego, un hongo como Physarum polycephalum o algoritmos computacionales pueden resolver laberintos optimizando el camino más corto.
  2. Propagación de Epidemias: Las epidemias pueden mostrar patrones similares de propagación en una población densa, donde la tasa de infección es alta en densidades críticas.
  3. Difusión de Contaminantes: La difusión de contaminantes en un medio fluido puede seguir patrones parecidos, donde la concentración y la tasa de dispersión afectan la propagación.

Se usó el automata 30: Según ChatGPT la regla 30 es una de las reglas más conocidas en el campo de los autómatas celulares unidimensionales, introducida por el matemático Stephen Wolfram. Es famosa por generar patrones complejos y caóticos a partir de reglas muy simples. En un autómata celular unidimensional, una línea de celdas puede estar en uno de dos estados: 0 (blanco) o 1 (negro). La evolución del autómata se realiza en pasos discretos de tiempo, comenzando con una configuración inicial que suele ser una línea con una sola celda en el estado 1 (negro) y todas las demás en el estado 0 (blanco).

Cada celda en el autómata celular 1D tiene una vecindad que incluye a sí misma y a sus dos vecinos inmediatos, formando un bloque de tres celdas. La regla 30 determina el estado de una celda en el siguiente paso de tiempo en función del estado de la celda y sus dos vecinos en el paso de tiempo actual. Existen ocho posibles combinaciones de estados para la vecindad de tres celdas, y cada combinación se asigna a un nuevo estado (0 o 1) según la regla 30.

Las combinaciones y sus resultados en la regla 30 son las siguientes: si la vecindad es 111, 110, o 101, el nuevo estado será 0; si la vecindad es 100, 011, 010, o 001, el nuevo estado será 1; y si la vecindad es 000, el nuevo estado será 0. El nombre "regla 30" proviene del hecho de que las salidas "00011110" se interpretan como el número binario 30 en decimal.

Supongamos que tenemos una línea de siete celdas con una sola celda en el estado 1 en el centro. La configuración inicial es "0001000". Aplicando la regla 30 a cada celda (excepto las de los extremos, que tienen vecinos faltantes), obtenemos la siguiente evolución paso a paso: "0001000", "0011100", "0110010", "1101111", "0011000".

La regla 30 es conocida por su comportamiento caótico y su capacidad de generar patrones complejos a partir de reglas simples. A pesar de la simplicidad de las reglas, la evolución del sistema puede mostrar una gran riqueza de estructuras y comportamientos, lo que la hace interesante tanto para la teoría de la computación como para la investigación en sistemas complejos y teoría del caos.

La regla 30, al igual que otros autómatas celulares, sirve como un ejemplo de cómo la complejidad puede surgir de reglas simples. Esto tiene implicaciones en muchos campos, incluyendo la física, la biología, la teoría de la computación y más, mostrando cómo patrones complejos y comportamientos emergentes pueden surgir de interacciones simples y locales.

Se programo el 30 que devuelve un archivo CSV que esta en la carpeta de las tareas y un segundo código que lo convierte en archivo STL que es el que sería interpretado por una impresora, este archivo también está en la carpeta de las tareas.

El código y la imagen de como quedaría el automata impreso serán mostrados a continuación. Se uso https://www.viewstl.com/ para la visualización del automata en 3D.

```python
import numpy as np

# Definir la regla 30
def regla_30(izquierda, centro, derecha):
    return izquierda ^ (centro | derecha)

# Inicializar la primera fila
ancho = 100
pasos = 50
estado_inicial = np.zeros(ancho, dtype=int)
estado_inicial[ancho // 2] = 1

# Matriz para almacenar los estados
estados = np.zeros((pasos, ancho), dtype=int)
estados[0] = estado_inicial

# Simular el autómata
for i in range(1, pasos):
    for j in range(1, ancho - 1):
        estados[i, j] = regla_30(estados[i - 1, j - 1], estados[i - 1, j], estados[i - 1, j + 1])

# Guardar los estados en un archivo (opcional)
np.savetxt("estados_automata.csv", estados, fmt="%d", delimiter=",")
```

```python
import os
import bpy
import csv

# Definir el tamaño de cada cubo
tamano_cubo = 1.0

# Cargar los estados desde el archivo CSV
with open("estados_automata.csv") as csvfile:
    reader = csv.reader(csvfile)
    estados = np.array([[int(celda) for celda in fila] for fila in reader])

# Función para crear un cubo en Blender
def crear_cubo(x, y, z):
    bpy.ops.mesh.primitive_cube_add(size=tamano_cubo, location=(x * tamano_cubo, y * tamano_cubo, z * tamano_cubo))

# Crear cubos para cada celda activa
for t, fila in enumerate(estados):
    for x, celda in enumerate(fila):
        if celda == 1:
            crear_cubo(x, t, 0)  # Usar el tiempo como la segunda dimensión espacial

# Verificar si el directorio de salida existe, si no, crearlo
output_dir = "output"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Guardar el modelo como un archivo STL
output_filepath = os.path.join(output_dir, "automata.stl")
bpy.ops.export_mesh.stl(filepath=output_filepath)
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d

# Coordenadas de las droguerías identificadas manualmente
droguerias = np.array([
    [283, 523],
    [290, 493],
    [271, 470],
    [250, 460],
    [240, 405],
    [238, 387],
    [382, 379],
    [414, 341],
    [226, 336],
    [366, 317],
    [181, 292],
    [247, 270],
    [382, 252],
    [352, 251],
    [259, 240],
    [245, 199],
    [242, 193],
    [239, 148],
    [300, 104],
    [291, 100]
])

# Crear el diagrama de Voronoi
vor = Voronoi(droguerias)

# Plotear el diagrama de Voronoi sobre el plano de la ciudad
fig, ax = plt.subplots()
ax.imshow(img, extent=[0, img.shape[1], 0, img.shape[0]])

# Ajustar el eje y porque las imágenes usualmente tienen el origen en la esquina superior izquierda
ax.set_ylim(ax.get_ylim()[::-1])

# Plotear el diagrama de Voronoi
voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='orange', line_width=2, point_size=10)

# Plotear las droguerías
ax.plot(droguerias[:, 0], droguerias[:, 1], 'ro', markersize=5)

# Etiquetar las droguerías
for i, (x, y) in enumerate(droguerias):
    ax.text(x, y, f'Droguería {i+1}', fontsize=9, ha='right')

# Configurar los ejes y mostrar el gráfico
ax.set_xlim(0, img.shape[1])
ax.set_ylim(0, img.shape[0])
ax.set_title('Diagrama de Voronoi de Droguerías sobre el Plano de la Ciudad')
ax.set_xlabel('Coordenada X')
ax.set_ylabel('Coordenada Y')
ax.grid(which='both', color='grey', linestyle='--', linewidth=0.5)

# Inclinar los valores del eje x en 90 grados
plt.xticks(rotation=90)

plt.show()
```

```python
# Crear una cuadrícula de puntos en el plano
x = np.linspace(0, img.shape[1], 100)
y = np.linspace(0, img.shape[0], 100)
xx, yy = np.meshgrid(x, y)
puntos = np.c_[xx.ravel(), yy.ravel()]

# Calcular la distancia mínima de cada punto a cualquier droguería
distancias = np.min(np.linalg.norm(puntos[:, None, :] - droguerias[None, :, :], axis=2), axis=1)
distancias = distancias.reshape(xx.shape)

# Definir un umbral para considerar un área como desatendida
umbral = 55

# Identificar las áreas desatendidas
areas_desatendidas = distancias > umbral

# Plotear el resultado
fig, ax = plt.subplots()
ax.imshow(img, extent=[0, img.shape[1], 0, img.shape[0]])
ax.set_ylim(ax.get_ylim()[::-1])
ax.contourf(xx, yy, areas_desatendidas, levels=[0, 0.5, 1], colors=['white', 'red'], alpha=0.5)
voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='orange', line_width=2, point_size=10)
ax.plot(droguerias[:, 0], droguerias[:, 1], 'ro', markersize=5)
for i, (x, y) in enumerate(droguerias):
    ax.text(x, y, f'Droguería {i+1}', fontsize=9, ha='right')
ax.set_xlim(0, img.shape[1])
ax.set_ylim(0, img.shape[0])
ax.set_title('Áreas Desatendidas por Droguerías sobre el Plano de la Ciudad')
ax.set_xlabel('Coordenada X')
ax.set_ylabel('Coordenada Y')
ax.grid(which='both', color='grey', linestyle='--', linewidth=0.5)

# Inclinar los valores del eje x en 90 grados
plt.xticks(rotation=90)

plt.show()
```

Comenzamos creando la simulación en Python y luego un código para contar las casillas en cada estado.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from ipywidgets import interact, IntSlider
import io
from PIL import Image

# Parámetros (puedes ajustarlos según sea necesario)
N = 100  # Tamaño del retículo (lattice)
ciclos = 100  # Número de ciclos de simulación
ps = 0.4  # Probabilidad de infección de una persona sana por vecinos sintomáticos
pa = 0.3  # Probabilidad de infección de una persona sana por vecinos asintomáticos
rs = 0.2  # Probabilidad de recuperación del estado sintomático en cada ciclo
ra = 0.1  # Probabilidad de recuperación del estado asintomático en cada ciclo
md = 0.1  # Probabilidad de muerte en estado sintomático en cada ciclo
NP, PS, CS, CNS, CT, PAC = range(6)  # Índices de estado para mayor claridad

def init(N, ns, cs, ca):
    """Inicializa el retículo con números específicos de personas."""
    g = np.zeros((N, N), int)
    for _ in range(ns):
        x, y = random.randint(0, N - 1), random.randint(0, N - 1)
        while g[x, y]:
            x, y = random.randint(0, N - 1), random.randint(0, N - 1)
        g[x, y] = PS  # Persona sana
    for _ in range(cs):
        x, y = random.randint(0, N - 1), random.randint(0, N - 1)
        while g[x, y]:
            x, y = random.randint(0, N - 1), random.randint(0, N - 1)
        g[x, y] = CS  # Persona con COVID-19 sintomática
    for _ in range(ca):
        x, y = random.randint(0, N - 1), random.randint(0, N - 1)
        while g[x, y]:
            x, y = random.randint(0, N - 1), random.randint(0, N - 1)
        g[x, y] = CNS  # Persona con COVID-19 asintomática
    return g

def step(g, muertos):
    """Simula un paso temporal, actualizando el estado de cada celda."""
    ng = g.copy()
    for x in range(N):
        for y in range(N):
            s = g[x, y]
            if s == PS:  # Persona sana
                n_infected = sum(g[(x + i) % N, (y + j) % N] in [CS, CNS]
                                 for i in [-1, 0, 1] for j in [-1, 0, 1] if (i, j) != (0, 0))
                if n_infected > 0:
                    if random.random() < ps:  # Mayor probabilidad de infección por sintomático
                        ng[x, y] = CS
                    elif random.random() < pa:  # Menor probabilidad de infección por asintomático
                        ng[x, y] = CNS
            elif s == CS:  # Persona con COVID-19 sintomática
                if random.random() < md:  # Muerte
                    ng[x, y] = NP
                    muertos += 1
                elif random.random() < rs:  # Recuperación
                    ng[x, y] = PAC
                else:  # Continúa sintomático
                    ng[x, y] = CS
            elif s == CNS:  # Persona con COVID-19 asintomático
                if random.random() < ra:  # Recuperación
                    ng[x, y] = PAC
                else:  # Continúa asintomático
                    ng[x, y] = CNS
            elif s == CT:  # En cuarentena o en tratamiento
                if random.random() < rs:
                    ng[x, y] = PAC  # Recupera
                else:
                    ng[x, y] = NP  # Muere
                    muertos += 1
            elif s == PAC:  # Persona curada con anticuerpos
                ng[x, y] = PAC  # Se asume que la inmunidad es persistente
    return ng, muertos

def calcular_porcentajes(g, muertos):
    """Calcula el porcentaje de cada estado en el retículo y el porcentaje de muertos."""
    unique, counts = np.unique(g, return_counts=True)
    total = g.size
    porcentajes = dict(zip(unique, counts / total * 100))
    porcentajes[6] = (muertos / total) * 100  # Añadir el porcentaje de muertos con índice 6
    return porcentajes

def almacenar_porcentajes(porcentajes):
    """Almacena los porcentajes de cada estado en una cadena formateada."""
    nombres_estados = ['No hay nada', 'Persona sana', 'COVID-19 sintomática', 'COVID-19 asintomática', 'Cuidados médicos', 'Curada con anticuerpos', 'Muertos']
    texto_porcentajes = "\n".join([f'{nombres_estados[estado]}: {porcentaje:.2f}%' for estado, porcentaje in porcentajes.items()])
    return texto_porcentajes

def sim():
    """Ejecuta la simulación completa."""
    g = init(N, 5000, 5, 5)
    imgs = []  # Lista para almacenar las imágenes de cada ciclo
    porcentajes_info = []  # Lista para almacenar los porcentajes de cada ciclo
    muertos = 0  # Contador de muertos

    for _ in range(ciclos):
        g, muertos = step(g, muertos)

        # Calcular los porcentajes de cada estado y de muertos
        porcentajes = calcular_porcentajes(g, muertos)
        # Almacenar la información de los porcentajes en el buffer
        porcentajes_info.append(almacenar_porcentajes(porcentajes))

        # Almacenar imágenes sin mostrarlas individualmente (mejora el rendimiento)
        buf = io.BytesIO()
        plt.imshow(g, cmap='viridis', vmin=0, vmax=5)
        plt.savefig(buf, format='png')
        buf.seek(0)
        imgs.append(Image.open(buf))
        plt.close()

    # Crear el control deslizante interactivo
    def show(i):
        # Mostrar la imagen correspondiente al ciclo i
        plt.imshow(imgs[i], cmap='viridis', vmin=0, vmax=5)
        # Mostrar la barra de color con etiquetas personalizadas equidistantes
        cbar = plt.colorbar(ticks=[0, 1, 2, 3, 4, 5])
        cbar.set_ticklabels(['No hay nada', 'Persona sana', 'COVID-19 sintomática', 'COVID-19 asintomática', 'Cuidados médicos', 'Curada con anticuerpos'])
        # Mostrar la información de los porcentajes
        plt.figtext(0.15, 0.85, porcentajes_info[i], fontsize=10, bbox={"facecolor": "white", "alpha": 0.5, "pad": 5})
        plt.show()

    interact(show, i=IntSlider(min=0, max=len(imgs) - 1))

sim()
```
